```python
import math
class Linear_Regression:
    # Initiliase the class with user defined learning rate and number of epochs
    def __init__(self, learning_rate, n_epochs):
        self.learning_rate = learning_rate
        self.n_epochs = n_epochs
        
    
    def gradient_descent(self,x,y):
        theta_curr = np.zeros(x.shape[1]) #weights
        bias = 0 #bias
        n_samples = y.size
        cost_l = [0] * self.n_epochs
        rmse = [0] * self.n_epochs

        for i in range(self.n_epochs):
            y_pred = x.dot(theta_curr) + bias
            
            #computing gradients
            dm = (-2/n_samples)*x.T.dot(y-y_pred)
            dc = (-2/n_samples)*np.sum(y-y_pred)

            #Updating weights
            theta_curr = theta_curr - self.learning_rate*dm
            bias = bias -  self.learning_rate*dc
            
            #Calculating the cost associated with each iteration
            cost = (1/(2*n_samples))*np.sum((y_pred-y)**2)
            cost_l[i] = cost 
            mse = ((1/(n_samples))*np.sum((y_pred-y)**2))
            mse = math.sqrt(mse)
            rmse[i] = mse
        return theta_curr, bias, cost_l, rmse
    
    def gradient_descent_l1(self,x,y,lambda_1):
        theta_curr_l1 = np.zeros(x.shape[1]) #weights
        bias_l1 = 0 #bias
        n_samples = y.size
        cost_l_l1 = [0] * self.n_epochs
        rmse_l1 = [0] * self.n_epochs

        for i in range(self.n_epochs):
            y_pred = x.dot(theta_curr_l1) + bias_l1
            
            #computing gradients
            dm = (-2/n_samples)*x.T.dot(y-y_pred)
            dc = (-2/n_samples)*np.sum(y-y_pred)

            #Updating weights
            theta_curr_l1 = theta_curr_l1 - self.learning_rate*dm
            bias_l1 = bias_l1 -  self.learning_rate*dc
            
            #Calculating the cost associated with each iteration
            cost = (1/(2*n_samples))*np.sum((y_pred-y)**2 + lambda_1*np.sum(np.power(theta_curr_l1,2)))
            cost_l_l1[i] = cost 
            mse = ((1/(n_samples))*np.sum((y_pred-y)**2))
            mse = math.sqrt(mse)
            rmse_l1[i] = mse
        return theta_curr_l1, bias_l1, cost_l_l1, rmse_l1
    
    def gradient_descent_l2(self,x,y,lambda_2):
        theta_curr_l2 = np.zeros(x.shape[1]) #weights
        bias_l2 = 0 #bias
        n_samples = y.size
        cost_l_l2 = [0] * self.n_epochs
        rmse_l2 = [0] * self.n_epochs

        for i in range(self.n_epochs):
            y_pred = x.dot(theta_curr_l2) + bias_l2
            
            #computing gradients
            dm = (-2/n_samples)*x.T.dot(y-y_pred)
            dc = (-2/n_samples)*np.sum(y-y_pred)

            #Updating weights
            theta_curr_l2 = theta_curr_l2 - self.learning_rate*dm
            bias_l2 = bias_l2 -  self.learning_rate*dc
            
            #Calculating the cost associated with each iteration
            cost = (1/(2*n_samples))*np.sum((y_pred-y)**2 + lambda_2*np.sum(np.absolute(theta_curr_l2)))
            cost_l_l2[i] = cost 
            mse = ((1/(n_samples))*np.sum((y_pred-y)**2))
            mse = math.sqrt(mse)
            rmse_l2[i] = mse
        return theta_curr_l2, bias_l2, cost_l_l2, rmse_l2
    
    
    def predict(self,X,theta,bias):
        return X.dot(theta) + bias
    
    
```

```python
model = Linear_Regression(learning_rate=0.01,n_epochs=200)

kf = KFold(n_splits=5)
Y = train['rings']
X = train.drop(['rings'],axis=1)

X = X.to_numpy()
Y = Y.to_numpy()

index = 0
rmse_sum = []
t_min = []

rmse_sum_l1 = []
t_min_l1 = []

rmse_sum_l2 = []
t_min_l2 = []

for train_index, test_index in kf.split(train): 
    X_train, X_test = X[train_index], X[test_index] 
    Y_train, Y_test = Y[train_index], Y[test_index]
    
    #t: Theta obtained after gradient descent, b: Bias obtained after gradient desecnt, c: cost value
    t,b,c,rmse = model.gradient_descent(X_train,Y_train)
    y_pred_train = X_train.dot(t)+b
    y_pred = model.predict(X_test,t,b) #validation set
#     fig, axes = plt.subplots(1, 3)
#     axes[0].plot(rmse,'g') 
    plt.plot(rmse)
    t_len = len(t)
    t_min.extend(t)
    rmse_sum.append(sum(rmse))


for train_index, test_index in kf.split(train): 
    X_train, X_test = X[train_index], X[test_index] 
    Y_train, Y_test = Y[train_index], Y[test_index]
     #t: Theta obtained after gradient descent, b: Bias obtained after gradient desecnt, c: cost value
    t_l1,b_l1,c_l1,rmse_l1 = model.gradient_descent_l1(X_train,Y_train,2)
    y_pred_train_l1 = X_train.dot(t_l1)+b_l1
    y_pred_l1 = model.predict(X_test,t_l1,b_l1) #validation set
#     axes[1].plot(rmse_l1, 'm')
    plt.plot(rmse_l1)
    t_len_l1 = len(t_l1)
    t_min_l1.extend(t_l1)
    rmse_sum_l1.append(sum(rmse_l1))
    

for train_index, test_index in kf.split(train): 
    X_train, X_test = X[train_index], X[test_index] 
    Y_train, Y_test = Y[train_index], Y[test_index]
     #t: Theta obtained after gradient descent, b: Bias obtained after gradient desecnt, c: cost value
    t_l2,b_l2,c_l2,rmse_l2 = model.gradient_descent_l2(X_train,Y_train,2)
    y_pred_train_l2 = X_train.dot(t_l2)+b_l2
    y_pred_l2 = model.predict(X_test,t_l2,b_l2) #validation set
#     axes[2].plot(rmse_l2, 'b')
    plt.plot(rmse_l2)
    t_len_l2 = len(t_l2)
    t_min_l2.extend(t_l2)
    rmse_sum_l2.append(sum(rmse_l2))
#     print(t)


min_ind = rmse_sum.index(min(rmse_sum))
t_best = t_min[min_ind*t_len:(min_ind*t_len+t_len)]
# t_best

min_ind_l1 = rmse_sum_l1.index(min(rmse_sum_l1))
t_best_l1 = t_min_l1[min_ind_l1*t_len_l1:(min_ind_l1*t_len_l1+t_len_l1)]
# t_best

min_ind_l2 = rmse_sum_l2.index(min(rmse_sum_l2))
t_best_l2 = t_min_l2[min_ind_l2*t_len_l2:(min_ind_l2*t_len_l2+t_len_l2)]
# t_best
```
